```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
def scrape_page(url):
    headers = headers = { 
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebkit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36",
    "Accept-Language": "en-US, en;q=0.9",
}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    products = soup.find_all("div", {"data-component-type": "s-search-result"})
    data = []
 
    for product in products:
       title = product.find("h2", class_="a-text-normal") 
       price = product.find("span", class_="a-offscreen")
       rating = product.find("span", class_="a-icon-alt")
       image = product.find("img")
       data.append([
          title.text.strip() if title else "N/A",
          price.text.strip() if price else "N/A",
          rating.text.strip() if rating else "N/A",
          image["src"] if image else "N/A"
      ])
    return data, soup

base_url = "https://www.amazon.com/s?k=laptops&crid=24KNS3N21HOLA&sprefix=laptops%2Caps%2C296&ref=nb_sb_noss_1"
all_data = []
max_pages = 20
current_page = 1
url = base_url

while current_page <= max_pages:
    print(f"Scraping page {current_page}...")
    page_data, soup = scrape_page(url)
    all_data.extend(page_data)
    next_page = soup.find("a", class_="s-pagination-next")
    if next_page and "href" in next_page.attrs:
        next_url = "https://www.amazon.com/s?k=laptops&crid=24KNS3N21HOLA&sprefix=laptops%2Caps%2C296&ref=nb_sb_noss_1" + next_page["href"]
        url = next_url
        current_page += 1
        time.sleep(5)
    else:
        break

df = pd.DataFrame(all_data, columns=["Title", "Price", "Ratings","Image URL"])

```

```python
import pandas as pd
import re
df = pd.read_csv(r"C:\Users\DELL\amazon_laptops2.csv")
df
```

```python
categories = {
    "Gaming": ["gaming", "RTX", "GEForce", "high-performance", "RGB"],
    "Office": ["business", "office", "lightweight", "professsional"],
    "Student": ["student", "budget", "Chromebook", "affordable"]
}

def categorize_laptop(title):
    title_lower = title.lower()
    for category, keywords in categories.items():
        if any(keyword in title_lower for keyword in keywords):
            return category
    return "Other"

df["Category"] = df["Title"].apply(categorize_laptop)
```

```python
def clean_title(title):
    if pd.isna(title):
        return ""

    title = re.sub(r"[^a-zA-Z0-9\s]", "", title)
    title = re.sub(r"\s+", " ",title).strip()
    title = title.title()
    return title

df["Cleaned_Title"] = df["Title"].apply(clean_title)
```

```python
def extract_ram(title):
    match = re.search(r'(\d{1,3}GB\s*(?:RAM/DDR\d)?)', str(title), re.IGNORECASE)
    return match.group(1) if match else None

df["RAM"] = df["Title"].apply(extract_ram)
```

```python
def extract_storage(title):
    match = re.search(r'(\d{1,4}GB|dTB)\s*(SSD|HDD|eMMC)', str(title), re.IGNORECASE)
    return match.group(0) if match else None

df["Storage"] = df["Title"].apply(extract_storage)
```

```python
df["Brand"] = df["Title"].apply(lambda x: x.split()[0])
```
